```python
import pandas as pd 
import numpy as np
data = 'data/data.csv'
data2 = 'data/data2.csv'
data = pd.read_csv(data)
data2 = pd.read_csv(data2)
```

```python
combined = pd.merge(data, data2, how="inner", on="Date")
combined
```

```python
mean = combined.groupby('Date')['Fatalities_y'].mean()
median = combined.groupby('Date')['Fatalities_y'].median()
var = combined.groupby('Date')['Fatalities_y'].var()
s_d = combined.groupby('Date')['Fatalities_y'].std()
sem = combined.groupby('Date')['Fatalities_y'].sem()
summary = pd.DataFrame({"Mean": mean, "Median": median, "Variance": var, "Standard Deviation": s_d, "SEM": sem})
summary
```

```python
combined['Fatalities_y'] = pd.to_datetime(combined['Date'], errors='coerce')
combined= combined.dropna(subset=['Date'])
combined.head()
```

```python
combined.drop('Registration', inplace=True, axis=1)
```
